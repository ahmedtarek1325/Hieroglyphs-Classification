```python
import pandas as pd
import camelot.io as camelot
```

```python
# File Source
Source = "./Material/- SIMPLE_HIEROGLYPHIC_DICTIONARY.pdf"

# extract all the tables in the PDF file
tables = camelot.read_pdf(Source , pages='1-440')   #address of file location

All_Data = []
for table in tables:
    All_Data.append(table.df)
    
All_Data = pd.concat(All_Data)
```

```python
# print the first table as Pandas DataFrame
print(((All_Data)).shape)
sub_df = All_Data[[1,2,3]]
sub_df.columns = ['Gardiner' , 'Transliteration' , 'Translation']
#sub_df.to_csv("./Gradiner_trans.csv" , index= False)
```
